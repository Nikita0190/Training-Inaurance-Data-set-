```python
from pycaret.datasets import get_data
from pycaret.regression import *

insuranceDataSet = get_data("insurance")    # SN is 46
setup(data = insuranceDataSet, target='charges')

gbrModel = create_model('gbr')
```

```python
newDataSet = get_data("insurance").iloc[:10]
```

```python
newPredictions = predict_model(gbrModel, data = newDataSet)
newPredictions
```

```python
import matplotlib.pyplot as plt

predicted = newPredictions.iloc[:,-1]     # Last column
actual = newPredictions.iloc[:,-2]        # 2nd last column

plt.scatter(actual, predicted)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Actul Vs Predicted')
plt.savefig("result-scatter-plot.jpg", dpi=300)
plt.show()
```

```python
gbrModelTuned = tune_model(gbrModel, n_iter=200)
```
